```julia
using JuMP, NLPModels, NLPModelsJuMP;
```

```julia
P = Model();
```

```julia
@variable(P, x[1:2]);
```

```julia
@NLobjective(P, Min, x[1]^2 + x[2]^2);
```

```julia
nlp = MathOptNLPModel(P);
```

```julia
x, f, gradnorm, iter, status, histf, histgradnorm = gradiente2(nlp,x0=[1;1]);
```

```julia
G = plot(histf,title="Função objetivo",xlabel="iter",label="f",size=(2800,1300),lw=10,ytickfont=font(40),xtickfont=font(40),titlefont=font(80),legendfont=font(40));
```
